```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from keras.models import Sequential, Model
from keras.layers import Conv2D, MaxPool2D, Dense, Dropout, BatchNormalization, GlobalAveragePooling2D, Input, Flatten
from keras.optimizers import Adam
from keras.applications import VGG16, MobileNet
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import tensorflow as tf
import random

# Define labels and image size
labels = ['Negative', 'Positive']
img_size = 100

# Function to read and process images
def read_images(data_dir):
    data = []
    labels_list = []
    for label in labels: 
        path = os.path.join(data_dir, label)
        class_num = labels.index(label)
        for img in os.listdir(path):
            try:
                img_arr = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
                resized_arr = cv2.resize(img_arr, (img_size, img_size)) 
                rgb_arr = np.stack([resized_arr] * 3, axis=-1)
                data.append(rgb_arr)
                labels_list.append(class_num)
            except Exception as e:
                print(f"Error reading image {img}: {e}")
    return np.array(data), np.array(labels_list)

# Load the dataset
x, y = read_images('/kaggle/input/surface-crack-detection')

# Preprocess data
x = x.astype("float32") / 255.
y = np.array(y)

# --- Display Sample Images (Before Model Setup) ---
def display_sample_images(data_dir):
    fig, axes = plt.subplots(1, 2, figsize=(8, 4))
    for idx, label in enumerate(labels):
        class_path = os.path.join(data_dir, label)
        sample_image = random.choice(os.listdir(class_path))
        img = cv2.imread(os.path.join(class_path, sample_image))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        axes[idx].imshow(img)
        axes[idx].set_title(f"Class: {label}")
        axes[idx].axis('off')
    plt.tight_layout()
    plt.show()

display_sample_images('/kaggle/input/surface-crack-detection')

# --- Custom CNN Model ---
cnn_model = Sequential([
    Conv2D(32, 3, padding="same", activation="relu", input_shape=(img_size, img_size, 3)),
    MaxPool2D(),
    Conv2D(64, 3, padding="same", activation="relu"),
    MaxPool2D(),
    GlobalAveragePooling2D(),
    Dense(64, activation="relu"),
    Dropout(0.5),
    BatchNormalization(),
    Dense(2, activation="softmax")
])
cnn_model.compile(loss="sparse_categorical_crossentropy", optimizer=Adam(1e-6), metrics=["accuracy"])
cnn_model.summary()

# --- VGG16 Transfer Learning Model ---
input_layer = Input(shape=(img_size, img_size, 3))
weights_path_vgg = r'/kaggle/input/dataset-vgg/vgg16_weights_tf_dim_ordering_tf_kernels_notop.h5'

if not os.path.exists(weights_path_vgg):
    print(f"Error: The file at {weights_path_vgg} does not exist!")
else:
    print(f"File found: {weights_path_vgg}")
    base_model_vgg = VGG16(weights=None, include_top=False, input_shape=(img_size, img_size, 3))
    base_model_vgg.load_weights(weights_path_vgg)
    for layer in base_model_vgg.layers:
        layer.trainable = False
    x_base_vgg = base_model_vgg(input_layer)
    x_base_vgg = GlobalAveragePooling2D()(x_base_vgg)
    x_base_vgg = Dense(128, activation='relu')(x_base_vgg)
    x_base_vgg = Dropout(0.5)(x_base_vgg)
    output_layer_vgg = Dense(2, activation='softmax')(x_base_vgg)
    vgg_model = Model(inputs=input_layer, outputs=output_layer_vgg)
    vgg_model.compile(loss="sparse_categorical_crossentropy", optimizer=Adam(1e-5), metrics=["accuracy"])
    vgg_model.summary()

# --- MobileNet Transfer Learning Model ---
input_layer = Input(shape=(img_size, img_size, 3))
weights_path_mobilenet = r'/kaggle/input/mobilenet/mobilenet_1_0_128_tf_no_top.h5'

if not os.path.exists(weights_path_mobilenet):
    print(f"Error: The file at {weights_path_mobilenet} does not exist!")
else:
    print(f"File found: {weights_path_mobilenet}")
    base_model_mobilenet = MobileNet(weights=None, include_top=False, input_shape=(img_size, img_size, 3))
    base_model_mobilenet.load_weights(weights_path_mobilenet)
    for layer in base_model_mobilenet.layers:
        layer.trainable = False
    x_base_mobilenet = base_model_mobilenet(input_layer)
    x_base_mobilenet = GlobalAveragePooling2D()(x_base_mobilenet)
    x_base_mobilenet = Dense(128, activation='relu')(x_base_mobilenet)
    x_base_mobilenet = Dropout(0.5)(x_base_mobilenet)
    output_layer_mobilenet = Dense(2, activation='softmax')(x_base_mobilenet)
    mobilenet_model = Model(inputs=input_layer, outputs=output_layer_mobilenet)
    mobilenet_model.compile(loss="sparse_categorical_crossentropy", optimizer=Adam(1e-5), metrics=["accuracy"])
    mobilenet_model.summary()

# --- Train Models ---
history_cnn = cnn_model.fit(x, y, epochs=5, batch_size=16, validation_split=0.25, verbose=1)
history_vgg = vgg_model.fit(x, y, epochs=2, batch_size=16, validation_split=0.25, verbose=1)
history_mobilenet = mobilenet_model.fit(x, y, epochs=2, batch_size=16, validation_split=0.25, verbose=1)

# --- Predictions ---
preds_cnn = np.argmax(cnn_model.predict(x), axis=1)
preds_vgg = np.argmax(vgg_model.predict(x), axis=1)
preds_mobilenet = np.argmax(mobilenet_model.predict(x), axis=1)

# --- Classification Reports FIRST ---
print("Classification Report - Custom CNN:")
print(classification_report(y, preds_cnn, target_names=labels))

print("Classification Report - VGG16:")
print(classification_report(y, preds_vgg, target_names=labels))

print("Classification Report - MobileNet:")
print(classification_report(y, preds_mobilenet, target_names=labels))

# --- Accuracy vs Epoch & Loss vs Epoch ---
plt.figure(figsize=(12, 12))
plt.style.use('ggplot')

plt.subplot(3, 2, 1)
plt.plot(history_cnn.history['accuracy'])
plt.plot(history_cnn.history['val_accuracy'])
plt.title('Accuracy of Custom CNN')
plt.ylabel('Accuracy', fontsize=12)
plt.xlabel('Epoch', fontsize=12)
plt.legend(['train accuracy', 'validation accuracy'], loc='lower right', prop={'size': 12})

plt.subplot(3, 2, 2)
plt.plot(history_cnn.history['loss'])
plt.plot(history_cnn.history['val_loss'])
plt.title('Loss of Custom CNN')
plt.ylabel('Loss', fontsize=12)
plt.xlabel('Epoch', fontsize=12)
plt.legend(['train loss', 'validation loss'], loc='best', prop={'size': 12})

plt.subplot(3, 2, 3)
plt.plot(history_vgg.history['accuracy'])
plt.plot(history_vgg.history['val_accuracy'])
plt.title('Accuracy of VGG16')
plt.ylabel('Accuracy', fontsize=12)
plt.xlabel('Epoch', fontsize=12)
plt.legend(['train accuracy', 'validation accuracy'], loc='lower right', prop={'size': 12})

plt.subplot(3, 2, 4)
plt.plot(history_vgg.history['loss'])
plt.plot(history_vgg.history['val_loss'])
plt.title('Loss of VGG16')
plt.ylabel('Loss', fontsize=12)
plt.xlabel('Epoch', fontsize=12)
plt.legend(['train loss', 'validation loss'], loc='best', prop={'size': 12})

plt.subplot(3, 2, 5)
plt.plot(history_mobilenet.history['accuracy'])
plt.plot(history_mobilenet.history['val_accuracy'])
plt.title('Accuracy of MobileNet')
plt.ylabel('Accuracy', fontsize=12)
plt.xlabel('Epoch', fontsize=12)
plt.legend(['train accuracy', 'validation accuracy'], loc='lower right', prop={'size': 12})

plt.subplot(3, 2, 6)
plt.plot(history_mobilenet.history['loss'])
plt.plot(history_mobilenet.history['val_loss'])
plt.title('Loss of MobileNet')
plt.ylabel('Loss', fontsize=12)
plt.xlabel('Epoch', fontsize=12)
plt.legend(['train loss', 'validation loss'], loc='best', prop={'size': 12})

plt.tight_layout()
plt.show()

# --- Confusion Matrices ---
def display_confusion_matrix(y_true, y_pred, model_name):
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f'Confusion Matrix - {model_name}')
    plt.show()

display_confusion_matrix(y, preds_cnn, "Custom CNN")
display_confusion_matrix(y, preds_vgg, "VGG16")
display_confusion_matrix(y, preds_mobilenet, "MobileNet")

# --- Random Prediction Demo ---
def predict_random_images(model, model_name, data_dir):
    print(f"\nPredictions from {model_name}:")
    for label in labels:
        class_path = os.path.join(data_dir, label)
        sample_image = random.choice(os.listdir(class_path))
        img_path = os.path.join(class_path, sample_image)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        img_resized = cv2.resize(img, (img_size, img_size))
        img_rgb = np.stack([img_resized] * 3, axis=-1) / 255.
        img_rgb = np.expand_dims(img_rgb, axis=0)
        prediction = np.argmax(model.predict(img_rgb), axis=1)[0]
        plt.imshow(cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB))
        plt.title(f"True: {label}, Predicted: {labels[prediction]}")
        plt.axis('off')
        plt.show()

predict_random_images(cnn_model, "Custom CNN", '/kaggle/input/surface-crack-detection')
predict_random_images(vgg_model, "VGG16", '/kaggle/input/surface-crack-detection')
predict_random_images(mobilenet_model, "MobileNet", '/kaggle/input/surface-crack-detection')
```
